```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
home=pd.read_csv('Housing.csv')
home.head()
```

```python
#Line Plot
import matplotlib.pyplot as plt
plt.plot(home['bedrooms'],home['guestroom'])
plt.title('line plot')
plt.xlabel('bedrooms')
plt.ylabel('guestroom')
plt.show()
```

```python
#Scatter Plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='bedrooms',y='bathrooms',data=home)
plt.title('Scatter plot')
plt.xlabel('bedrooms')
plt.ylabel('bathrooms')
plt.show()
```

```python
#Count Plot
plt.figure(figsize=(15,5))
sns.countplot(y='bedrooms',data=home)
```

```python
#Heat Map
data=np.random.rand(10,10)
sns.heatmap(data)
```

```python
#Histogram
import matplotlib.pyplot as plt
plt.hist(home['area'],
         bins=20,
         edgecolor='black')
plt.title('Histogram')
plt.xlabel('area')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box Plot
sns.boxplot(x='bedrooms',y='area',data=home,width=0.5)
plt.title('Box Plot')
plt.xlabel('bedrooms')
plt.ylabel('area')
plt.show()
```

```python
#Pie Chart

stories_counts = home['stories'].value_counts()
plt.pie(stories_counts, labels=stories_counts.index)#, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()

home.head()
```
